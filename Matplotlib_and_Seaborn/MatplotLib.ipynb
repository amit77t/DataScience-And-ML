```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# MatplotLib Basic

```python
x=np.linspace(0,5,11)
x
y=x**2
y
```

```python
plt.plot(x,y)
plt.title("Hello")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

# Subplot in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x*x,y)
```

# Object Oriented Way

```python
fig=plt.figure()
axis1=fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.0,0.4,.3,.3])
axis2.plot(x,y)
```

## Types of plot

```python
fig=plt.figure(figsize=(12,8) ,dpi=100)
axis1=fig.add_axes([0.1, 0.1, 0.4, 0.8])
axis1.plot(x,y)
axis2=fig.add_axes([0.3,0.4,.3,.3])
axis2.plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.boxplot(x)
```

## Saving your plot


```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,.4,.4])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
