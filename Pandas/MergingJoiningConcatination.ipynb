```python
import numpy as np
import pandas as pd
```

## Merging 2 DataFrame




```python
employee_info = {
    "employeeId": [101, 102, 103, 104, 105],
    "name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "department": ["HR", "Engineering", "Marketing", "Sales", "Finance"]
}
salary_info = {
    "employeeId": [101, 102, 106, 107, 105],  # 106 and 107 are not in employee_info
    "salary": [50000, 80000, 72000, 65000, 70000],
    "bonus": [5000, 8000, 7200, 6500, 7000]
}

df1=pd.DataFrame(employee_info)
df2=pd.DataFrame(salary_info)
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on="employeeId", how='inner')
```

```python
pd.merge(df1, df2, on="employeeId", how='outer')
```

```python
pd.merge(df1, df2, on="employeeId", how='left')
```

```python
pd.merge(df1, df2, on="employeeId", how='right')
```

## Concatination of 2 DataFrame
 

```python


df1 = pd.DataFrame({
    "employeeId": [101, 102, 103],
    "name": ["Alice", "Bob", "Charlie"],
    "department": ["HR", "Engineering", "Marketing"]
})
df2 = pd.DataFrame({
    "employeeId": [104, 105],
    "name": ["David", "Eva"],
    "department": ["Sales", "Finance"]
    
})
df1
df2
```

```python

pd.concat([df1,df2])
```

```python
pd.concat([df1, df2], axis=1)
```
