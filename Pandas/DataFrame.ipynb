## Creating a DataFrame

```python
import numpy as np
import pandas as pd
```

```python
data = {
    "name": ["Amit", "Priya", "Rohan", "Ananya", "Vikram"],
    "age": [30, 28, 35, 26, 40],
    "city": ["Allahabad", "Mumbai", "Delhi", "Bengaluru", "Chennai"],
    "salary": [55000, 68000, 72000, 64000, 80000]
}
df=pd.DataFrame(data)
df
```

```python
people = [
    ["Aarav Mehta", 29, 75000, "Mumbai"],
    ["Sara Kapoor", 34, 92000, "Delhi"],
    ["Rohan Das", 41, 105000, "Bangalore"],
    ["Neha Verma", 26, 68000, "Hyderabad"],
    ["Kabir Singh", 38, 99000, "Chennai"]
]

```

```python
df2=pd.DataFrame(people)
```

```python
df2
```

```python
columns=["Name", "Age", "Salary", "City"]
df2=pd.DataFrame(people, columns=columns)
df2
```
