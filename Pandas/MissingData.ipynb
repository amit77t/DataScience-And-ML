```python
import numpy as np
import pandas as pd
```

## Finding Missing Data

```python


data = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Priya', 'Karan', 'Sana', 'Vikram', 'Anita'],
    'Age': [28, np.nan, 35, 42, np.nan, 30, 27, np.nan],
    'City': ['Delhi', 'Mumbai', 'Delhi', None, 'Bangalore', 'Delhi', None, 'Mumbai'],
    'Income': [50000, 62000, None, 72000, 55000, None, 47000, 65000]
}

df2 = pd.DataFrame(data)
df2
```

```python
df2.isna().sum()
```

```python
df2.isna().any()
```

## Removing Missing Value

```python
df2
```

```python
df2.dropna()
```

```python
df2
```

```python
df2.dropna(thresh=1)
```

## Filling Null Value

```python
df2
```

```python
df2.fillna(0)
```

```python
values={"Name":"Unknown", "Age": 28, "City":"Pune", "Income":590894}  ##column wise
df2.fillna(value=values)
```

```python
df2
```

```python
fill_values = {
    'Age': df2['Age'].median(),           # numeric: median
    'City': 'Unknown',                     # categorical: constant
    'Income': df2['Income'].mean()         # numeric: mean
}
df2.fillna(value=fill_values)
```
